```python
# Read the CSV map of index names and paths and make a Pandas DataFrame
df = pd.read_csv('map.csv')
```

```python
# Make a Python dictionary from the DataFrame
map_dict = dict(zip(df['index'], df['path']))
# Preview the dictionary
map_dict
```

The next cell does the following:

- Established the base path for the site
- Walks the PATH and gets all the file and directory names
- Filters the files to work on just HTML
- Opens each HTML file and reads the content
- Finds for each "old" (`index`) pattern and replaces it with a new one (`path`)
- Writes the result

```python
PATH = '/Users/sjhuskey/Sites/create-projects/montealban.oucreate.com'
for dirpath, dirnames, filenames in os.walk(PATH):
    for filename in filenames:
        if filename.endswith('.html'):
            filepath = os.path.join(dirpath, filename)
            with open(filepath, 'r', encoding='utf-8') as file:
                content = file.read()
                # Perform string replacements based on map_dict
                for old, new in map_dict.items():
                    content = content.replace(old, new)
            with open(filepath, 'w', encoding='utf-8') as file:
                file.write(content)
```
