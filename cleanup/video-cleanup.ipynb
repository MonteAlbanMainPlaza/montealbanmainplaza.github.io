```python
import re
from pathlib import Path
```

```python
ROOT = Path("/Users/sjhuskey/Sites/create-projects/montealban.oucreate.com")  # <-- CHANGE THIS

# Match iframe src that points to YouTube embed, capturing the 11-char video id
YOUTUBE_IFRAME_SRC_RE = re.compile(
    r'(?P<prefix>\bsrc=["\'])'
    r"(?P<url>(?:https?:)?//(?:www\.)?youtube\.com/embed/(?P<vid>[A-Za-z0-9_-]{11}))"
    r'(?P<query>\?[^"\']*)?'
    r'(?P<suffix>["\'])',
    re.IGNORECASE,
)

changed_files = 0
changed_links = 0

for html_path in ROOT.rglob("*.html"):
    text = html_path.read_text(encoding="utf-8", errors="ignore")

    def repl(m: re.Match) -> str:
        nonlocal_changed = 1  # dummy to keep it obvious; not used
        # Build a clean embed URL without query params
        vid = m.group("vid")
        new_url = f"https://www.youtube-nocookie.com/embed/{vid}"
        return f'{m.group("prefix")}{new_url}{m.group("suffix")}'

    new_text, n = YOUTUBE_IFRAME_SRC_RE.subn(repl, text)

    if n:
        html_path.write_text(new_text, encoding="utf-8")
        changed_files += 1
        changed_links += n

print(
    f"Done. Updated {changed_links} iframe src URL(s) across {changed_files} file(s)."
)
```
